```python
import pandas as pd

df = pd.read_csv(u"carpetas-de-investigacion-pgj-cdmx.csv", nrows = 70000000,sep=';')
df.dropna()
print(df.describe())
print(df.info())
```

Nuestro dataframe cuenta con un total de 766303 entradas, los valores más importantes para nuestro análisis es la fecha del delito, la categoría del delito, la ubicación geográfica y la Alcaldía.

El primer análisis que realizamos tiene que ver con identificar la tendencia general de los crímenes.

```python
import numpy as np
import matplotlib.pyplot as plt

timedf = df[["Mes y año","Alcaldía","Geopoint"]].dropna()
timedf.loc[timedf["Mes y año"] == "01/2019", 'Mes y año'] = "2019-01"
timedf.loc[timedf["Mes y año"] == "02/2019", 'Mes y año'] = "2019-02"
timedf.loc[timedf["Mes y año"] == "feb-19", 'Mes y año'] = "2019-02"
timedf = timedf.set_index(["Mes y año","Geopoint"])
timedf = timedf.count(level="Mes y año")
timedf.plot()
```

```python
from statsmodels.tsa.arima_model import ARMA
import statsmodels.api as sm
import warnings
import itertools

warnings.filterwarnings('ignore')


p = range(10)
q = range(10)

#pq = list(itertools.product(p,q))
pq = [[0,i] for i in range(10)]

for i in pq:
    try:
        mod = ARMA(timedf, order = i)
        results = mod.fit()
    except Exception as e:
        continue
        

timedf.index = pd.DatetimeIndex(timedf.index)
a = pd.to_datetime('2019-01')
print(a)
selected = timedf.index < a
timedf = timedf[selected]
#dates = pd.date_range( start = "2016-01-01" ,  end  =  "2018-12-01" )
#timedf = timedf.loc[dates]
#timedf.dropna(axis='rows')

print(timedf)

mod = ARMA(timedf, order = [1,5])
results = mod.fit()
pred_future = results.predict(start=pd.to_datetime('2016-01-01'), end=pd.to_datetime('2025-01-01'), dynamic=False)

pred_future2 = results.predict(start=pd.to_datetime('2017-01-01'), end=pd.to_datetime('2017-12-01'), dynamic=False)
pred_future3 = results.predict(start=pd.to_datetime('2018-01-01'), end=pd.to_datetime('2018-12-01'), dynamic=False)

timedf.plot(figsize=(15,6),label='datos')
pred_future.plot(label='predict')
plt.figure()
pred_future2.plot(label='predict')
plt.figure()
pred_future3.plot(label='predict')
#plt.ylabel('Numero de delitos')

#L=plt.legend()
#L.get_texts()[0].set_text('real')
```

Como se puede observar, la ocurrencia del delito va en aumento en lo últimos años, sin embargo esta información no nos aporta ningún criterio para establecer un plan de acción.
